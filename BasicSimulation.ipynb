```julia
# meta parameters:
#n: number of students
#m: number of market rounds. A market round is defined as a buying, earning process.
#t: number of questions a student is allowed to bid, per market round
#q: number of questions a student creates, per round. In real life scenario, the number
# q is not fixed, but rather students have to pay to create questions. 
#w: vector of students quality 
#p: price of a question. Each person is deducted from its bank this ammount.
#k: number of people correctly answering a question
#l: number of people taking a question.
# to answer a question, tipically, p= 0. 
```

```julia
#Function to create a matrix containing the probability P(i>j) of student
#i answering correctly a question made by j. We use the RLT model, and 
#for this function we asume no dependance of question quality: answering
# a question well or not is simply a function of the student quality w.
#@param w: vector of students quality, vector has length n.
function createAnswerProbabilitiesRLT(w)
    dim= length(w)
    Q = zeros(Float64,(dim, dim))
    for i in 1:dim
        for j in 1: dim
          Q[i,j]= 1/(1+exp(-(w[i]-w[j])))  
        end
    end
    return (Q)
end
```

```julia
# Reward functions. All functions must recieve 2 parameters for consistency

# function to asign the reward to creator of a function with the 
#number of people that wrongly answered the question.
#@param l: number of people correctly answering the question
#@param k:
function rewardToCreatorWithWrongAnswers(ans)
     reward  =  sum(ans.==-1,dims=1)[1,:]
    return(reward)
end

# function to asign the reward to creator of a function with the 
#number of people that wrongly answered the question.
#@param l: number of people correctly answering the question
#@param k:
function rewardToCreatorWithNothing(ans)
    return(zeros(size(ans)))
end


```

```julia
#function to do 1 round of market. Each person selects
# a question randomly and bets on it, paying an amount p.
#@param p: cost of a question.
#@param n: the number of students.
#@parap rltProbas: matrix of probabilities given by the RLT
#@param K: number of questions a student is allowed to bid, per market round.
#param rewardFunction: function that gives the earnings to the maker of a question.
# in this particular implementation, K is fixed to 1. See variable responded well.
#@returns: payments: an array of size n containing how much each student earns.
#@returns: 
#CARE! in this function, a person can only bid on 1 question. if the person answers
#well or not is kept on the vector respondedWell. FIX K=1
function marketPhase(n,p,t,rltProbas,rewardFunction::Function)
    # each person bids on K questions. Recall i cannot bid on question made by him.
     keepCountSelected = zeros(n)
     keepCorrectlyAnswered =  zeros(n)
     respondedWell = zeros(n)
    earnings = zeros(n)
    for i in 1:n
          #  questionSelected =  sample(deleteat!(collect(1:n), i),k,replace=true)
         questionSelected =  Int(sample(deleteat!(collect(1:n), i),1)[1])
         keepCountSelected[questionSelected] =  keepCountSelected[questionSelected]+1
         
        succeed = 0+ rand(Bernoulli(rltProbas[i,questionSelected]),1)[1]
        if (succeed==1)
               respondedWell[i] = questionSelected
        end
        keepCorrectlyAnswered[questionSelected] = keepCorrectlyAnswered[questionSelected]+succeed
    end
    #payments[i] stores the ammount of money to be paid to people who correctly answered question[i].
    payments = (  keepCountSelected-keepCorrectlyAnswered)./keepCorrectlyAnswered
    
    # Deal with NaN: 0/0: no one took the question
    # Deal with inf : a/0 where a>1.
     for i in 1:n
        print
       if(isnan(payments[i]))
            payments[i]=0
            elseif (payments[i]==Inf)
                payments[i]=0
       end
   end
    #asign payments, adding rewards for good questions
    for i in 1:n
       
        if(respondedWell[i]!=0)
            earnings[i] =  payments[Int(respondedWell[i])]
        end
        earnings[i]= earnings[i]+ rewardFunction(keepCountSelected[i],keepCorrectlyAnswered[i])
    end
    
    
   # return(payments)
    return(earnings,payments,keepCountSelected,keepCorrectlyAnswered,respondedWell)
    #TODO: extend this function to the case where a person can ask multiple questions
    # and to the case where selecting questions is not random 
end
```

```julia
function marketPhaseMultiAnswers(n,p,t,rltProbas,rewardFunction::Function)
    # each person bids on K questions. Recall i cannot bid on question made by him.
     
    answers = zeros(n,n)
    
    earnings = zeros(n)
    for i in 1:n
          #  questionSelected =  sample(deleteat!(collect(1:n), i),k,replace=true)
         questionSelected =  Int.(sample(deleteat!(collect(1:n), i),t,replace=false))
         for j in questionSelected
            
        succeed = 0+ rand(Bernoulli(rltProbas[i,j]),1)[1]
            
           if (succeed==1)
              answers[i,j]=1
            else
                answers[i,j]=-1
            end        
        
       end
    end
    #payments[i] stores the ammount of money to be paid to people who correctly answered question[i].
    keepCorrectlyAnswered  =  sum(answers.==1,dims=1)[1,:]
    payments = (sum(abs.(answers),dims=1)[1,:]-keepCorrectlyAnswered)./keepCorrectlyAnswered
    # payments = 1 ./keepCorrectlyAnswered
    rewardsForQuestion = rewardFunction(answers)
    # Deal with NaN: 0/0: no one took the question
    # Deal with inf : a/0 where a>1.
     for i in 1:n
        
       if(isnan(payments[i]))
            payments[i]=0
            elseif (payments[i]==Inf)
                payments[i]=0
       end
   end
         

    #asign payments, adding rewards for good questions
    for i in 1:n
       
        for j in 1:n 
                if(answers[i,j]==1)
                   earnings[i] =  earnings[i]+payments[j]
                end
        end
            
        earnings[i]= earnings[i]+rewardsForQuestion[i]
    end
    
   # return(payments)
    return(earnings,payments,answers)
    #TODO: extend this function to the case where a person can ask multiple questions
    # and to the case where selecting questions is not random 
end
```
